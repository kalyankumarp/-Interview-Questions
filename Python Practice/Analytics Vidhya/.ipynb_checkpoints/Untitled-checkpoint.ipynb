```python
# importing required libraries
import seaborn as sns
sns.set()
sns.set(style="darkgrid")

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=(10,5)

import numpy as np
import pandas as pd
```

```python
# read the dataset
data_BM = pd.read_csv('bigmart_data.csv')
# drop the null values
data_BM = data_BM.dropna(how="any")
# multiply Item_Visibility by 100 to increase size
data_BM["Visibility_Scaled"] = data_BM["Item_Visibility"] * 100
# view the top results
data_BM.head()
```

```python
sns.lineplot('Item_Weight', 'Item_MRP', data = data_BM[:50])
```

```python

sns.relplot('Item_Weight', 'Item_MRP', data = data_BM[:50], kind = 'scatter')
```

```python
sns.barplot(x = 'Item_Type', y = 'Item_MRP', data = data_BM[:5])
```

```python
sns.distplot(data_BM['Item_MRP'])
```

```python
sns.distplot(data_BM['Item_MRP'], kde = False)
```

```python
sns.boxplot(y = 'Item_MRP', x = 'Item_Fat_Content', data = data_BM)
```

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", data=data_BM[:200], kind="scatter", hue="Visibility_Scaled", size="Visibility_Scaled");
```

```python
sns.relplot(x = 'Item_MRP', y = 'Item_Outlet_Sales', data = data_BM[:200], hue = 'Visibility_Scaled', size = 'Visibility_Scaled')
```

```python
sns.relplot(x = 'Item_MRP', y = 'Item_Outlet_Sales', data = data_BM[:200], col = 'Outlet_Size')
```

```python
sns.relplot(x = 'Item_MRP', y = 'Item_Outlet_Sales', data = data_BM[:200], col = 'Outlet_Size', style = 'Outlet_Size')
```
